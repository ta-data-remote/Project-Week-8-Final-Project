# Importing libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Import data

```python
df = pd.read_csv('./data/train.csv')
df2 = pd.read_csv('./data/train_cleaned_knn_imputation.csv')

titanic = df.copy()
titanic_new = df2.copy()
```

```python
titanic['Survived'].value_counts().sum()
```

```python
labels = titanic['Survived'].value_counts().index
shape = titanic['Survived'].value_counts().values

plt.figure(figsize=(12,8))
plt.pie(shape, labels=shape, autopct = '%1.1f%%', shadow=True)
plt.title('Survived', fontsize=20)
plt.style.use('default')
plt.legend(labels)

plt.show()
```

# How age affects survival rate(SR)?

![frequency_distribution_age](images/frequency_age.png)

```python
# kdeplot of survived by age

plt.figure(figsize=(12,8))
g = sns.kdeplot(titanic_new["Age"][(titanic_new["Survived"] == 0)], color="Red", shade=True)
g = sns.kdeplot(titanic_new["Age"][(titanic_new["Survived"] == 1)], ax=g, color="Blue", shade=True)
g.set_xlabel("Age")
g.set_ylabel("Frequency")
g.set_title('Frequency of survived and not survived by age')
g = g.legend(["Not Survived","Survived"])
```

# How sex affects SR?

```python
# barplot of survived by sex

plt.figure(figsize=(12,8))
ax = sns.barplot(x="Sex", y="Survived", data=titanic)
ax.set_ylabel('% Survived')
ax.set_xlabel('Gender')
ax.set_title("Survival rate by gender")
plt.show()


# survived rate by sex
dead_women = 81
dead_men = 468

survived_women = 233
survived_men = 109

total_women = dead_women + survived_women 
total_men = dead_men + survived_men
percentage_to_survive_if_women = ( survived_women / total_women ) * 100
percentage_to_survive_if_men = ( survived_men / total_men ) * 100

print(f'Chance to survive if you are woman: {percentage_to_survive_if_women:.2f} %')
print(f'Chance to survive if you are man: {percentage_to_survive_if_men:.2f} %')

# pivot table
print()

pd.pivot_table(titanic, index='Survived', columns='Sex', values='Ticket', aggfunc='count')
```

# How passenger's class affects SR?

```python
# barplot of survived by passenger's class
plt.figure(figsize=(12,8))
ax = sns.barplot(x="Pclass", y="Survived", data=titanic)
ax.set_ylabel('% Survived')
ax.set_xlabel('Passenger\'s class')
ax.set_title('Survival rate by passenger\'s class')
plt.show()

# survived rate by passenger's class
dead_first_class = 80
dead_second_class = 97
dead_third_class = 372

survived_first_class = 136 
survived_second_class = 87
survived_third_class = 119

total_first_class = dead_first_class + survived_first_class
total_second_class = dead_second_class + survived_second_class
total_third_class = dead_third_class + survived_third_class

percentage_to_survive_if_first_class = ( survived_first_class / total_first_class ) * 100
percentage_to_survive_if_second_class = ( survived_second_class / total_second_class ) * 100
percentage_to_surrvive_if_third_class = ( survived_third_class / total_third_class ) * 100


print(f'Chance to survive if you stayed in first class: {percentage_to_survive_if_first_class:.2f} %')
print(f'Chance to survive if you stayed in second class: {percentage_to_survive_if_second_class:.2f} %')
print(f'Chance to survive if you stayed in third class: {percentage_to_surrvive_if_third_class:.2f} %')


# pivot table

pd.pivot_table(titanic, index='Survived', columns='Pclass', values='Ticket', aggfunc='count')
```

# How do siblings and spouses affect SR?

```python
# barplot of survived by siblings and spouses


plt.figure(figsize=(12,8))
ax = sns.barplot(x="SibSp", y="Survived", data=titanic)
ax.set_ylabel('% Survived')
ax.set_xlabel('Number of siblings or spouses')
ax.set_title('Survival rate by siblings or spouses')
plt.show()

# survived rate by passenger's siblings or spouses
dead_without_siblings_or_spouses = 398
dead_having_1_siblings_or_spouses = 97
dead_having_2_siblings_or_spouses = 15
dead_having_3_siblings_or_spouses = 12
dead_having_4_siblings_or_spouses = 15
dead_having_5_siblings_or_spouses = 5
dead_having_8_siblings_or_spouses = 7

survived_without_siblings_or_spouses = 210
survived_having_1_siblings_or_spouses = 112
survived_having_2_siblings_or_spouses = 13
survived_having_3_siblings_or_spouses = 4
survived_having_4_siblings_or_spouses = 3
survived_having_5_siblings_or_spouses = 0
survived_having_8_siblings_or_spouses = 0

total_without_siblings_or_spouses = dead_without_siblings_or_spouses + survived_without_siblings_or_spouses
total_having_1_siblings_or_spouses = dead_having_1_siblings_or_spouses + survived_having_1_siblings_or_spouses
total_having_2_siblings_or_spouses = dead_having_2_siblings_or_spouses + survived_having_2_siblings_or_spouses
total_having_3_siblings_or_spouses = dead_having_3_siblings_or_spouses + survived_having_3_siblings_or_spouses
total_having_4_siblings_or_spouses = dead_having_4_siblings_or_spouses + survived_having_4_siblings_or_spouses
total_having_5_siblings_or_spouses = dead_having_5_siblings_or_spouses + survived_having_5_siblings_or_spouses
total_having_8_siblings_or_spouses = dead_having_8_siblings_or_spouses + survived_having_8_siblings_or_spouses


percentage_to_survive_if_without_siblings_or_spouses = ( survived_without_siblings_or_spouses / total_without_siblings_or_spouses ) * 100
percentage_to_survive_if_having_1_siblings_or_spouses = ( survived_having_1_siblings_or_spouses / total_having_1_siblings_or_spouses ) * 100
percentage_to_survive_if_having_2_siblings_or_spouses = ( survived_having_2_siblings_or_spouses / total_having_2_siblings_or_spouses ) * 100
percentage_to_survive_if_having_3_siblings_or_spouses = ( survived_having_3_siblings_or_spouses / total_having_3_siblings_or_spouses ) * 100
percentage_to_survive_if_having_4_siblings_or_spouses = ( survived_having_4_siblings_or_spouses / total_having_4_siblings_or_spouses ) * 100
percentage_to_survive_if_having_5_siblings_or_spouses = ( survived_having_5_siblings_or_spouses / total_having_5_siblings_or_spouses ) * 100
percentage_to_survive_if_having_8_siblings_or_spouses = ( survived_having_8_siblings_or_spouses / total_having_8_siblings_or_spouses ) * 100

print(f'Chance to survive if you travel without any siblings or spouses: {percentage_to_survive_if_without_siblings_or_spouses:.2f} %')
print(f'Chance to survive if you travel with 1 siblings or spouses: {percentage_to_survive_if_having_1_siblings_or_spouses:.2f} %')
print(f'Chance to survive if you travel with 2 siblings or spouses: {percentage_to_survive_if_having_2_siblings_or_spouses:.2f} %')
print(f'Chance to survive if you travel with 3 siblings or spouses: {percentage_to_survive_if_having_3_siblings_or_spouses:.2f} %')
print(f'Chance to survive if you travel with 4 siblings or spouses: {percentage_to_survive_if_having_4_siblings_or_spouses:.2f} %')
print(f'Chance to survive if you travel with 5 siblings or spouses: {percentage_to_survive_if_having_5_siblings_or_spouses:.2f} %')
print(f'Chance to survive if you travel with 8 siblings or spouses: {percentage_to_survive_if_having_8_siblings_or_spouses:.2f} %')

    

# pivot table

pd.pivot_table(titanic, index='Survived', columns='SibSp', values='Ticket', aggfunc='count')
```

# How port of embarkation affects SR?

```python
# barplot of survived by embarkation
plt.figure(figsize=(12,8))
ax = sns.barplot(x="Embarked", y="Survived", data=titanic)
ax.set_ylabel('% Survived')
ax.set_title('Survival rate by port of embarkation', size=12)
plt.show()

# survived rate by embarkation
dead_embarked_C = 75
dead_embarked_Q = 47
dead_embarked_S = 427

survived_embarked_C = 93
survived_embarked_Q = 30
survived_embarked_S = 219

total_embarked_C = dead_embarked_C + survived_embarked_C
total_embarked_Q = dead_embarked_Q + survived_embarked_Q
total_embarked_S = dead_embarked_S + survived_embarked_S

percentage_to_survive_if_embarked_C = ( survived_embarked_C / total_embarked_C ) * 100
percentage_to_survive_if_embarked_Q = ( survived_embarked_Q / total_embarked_Q ) * 100
percentage_to_survive_if_embarked_S = ( survived_embarked_S / total_embarked_S ) * 100


print(f'Chance to survive if embarked from Cherbourg: {percentage_to_survive_if_embarked_C:.2f} %')
print(f'Chance to survive if embarked from Queenstown: {percentage_to_survive_if_embarked_Q:.2f} %')
print(f'Chance to survive if embarked from Southampton: {percentage_to_survive_if_embarked_S:.2f} %')


## pivot table

pd.pivot_table(titanic, index='Survived', columns='Embarked', values='Ticket', aggfunc='count')
```
